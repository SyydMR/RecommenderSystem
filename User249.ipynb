```python
import pandas as pd
import numpy as np 
```

```python
dataset = pd.read_csv('ml-latest-small/ratings.csv')
movies = pd.read_csv('ml-latest-small/movies.csv')
```

```python
dataset
```

```python
from keras.models import load_model
model = load_model('user249_model.h5')
```

```python
for users in dataset.values:
    if int(users[0]) == 249:
        print(int(users[0]), int(users[1]), int(users[2]), int(users[3]))
```

```python
predictions = model.predict([dataset.userId.head(10), dataset.movieId.head(10)])
print(predictions)
```

```python
user_id = 249
u_users = len(dataset.userId.unique())
n_movies = len(dataset.movieId.unique())

movie_ids = np.arange(n_movies).reshape(-1, 1)
user_ids = np.full_like(movie_ids, user_id)

predicted_ratings = model.predict([user_ids, movie_ids])

dot_products = predicted_ratings.flatten()

sorted_indices = np.argsort(dot_products)[::-1]

top_movie_ids = sorted_indices[:10]

recommended_movie_ids = list(top_movie_ids)
recommended_movie_ids
```

```python
for movie_id in recommended_movie_ids:
    movie_row = movies[movies['movieId'] == movie_id]
    if not movie_row.empty:
        print(movie_row['title'].values[0])
```
